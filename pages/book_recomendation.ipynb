## Picking Out the books poster

```python
book_image=df1[["Book-Title","Image-URL-M"]]
book_image.rename(columns = {'Book-Title':'Book_Title'}, inplace = True) 
book_image.rename(columns = {'Image-URL-M':'Image'}, inplace = True) 
book_image
```

## Data Cleaning

```python
df1.isnull().sum()
df1.dropna(inplace=True)
df1.isnull().sum()
```

## Taking only Ratings and books csv

```python
df=df1.merge(df2,on="ISBN").drop_duplicates('Book-Title')
df.head()
```

## Selecting the required columns

```python
df=df[["Book-Title"	,"Book-Author",	"Year-Of-Publication",	"Publisher", "Book-Rating"]]
df
```

```python
df.rename(columns = {'Book-Title':'Book_Title'}, inplace = True) 
```

## Int to String conversion

```python
df["Year-Of-Publication"]=df["Year-Of-Publication"].apply(str)
df["Book-Rating"]=df["Book-Rating"].apply(str)
df
```

```python
df["Book-Author"]=df["Book-Author"].values.astype(str).tolist()
df["Publisher"]=df["Publisher"].values.astype(str).tolist()
df
```

```python
df["Book-Author"]=df["Book-Author"].str.replace(" ","")
df["Publisher"]=df["Publisher"].str.replace(" ","")
df
```

# Merging the text

```python
df['tag']=df["Book-Author"].apply(lambda x:x.split())+df["Year-Of-Publication"].apply(lambda x:x.split())+df["Publisher"].apply(lambda x:x.split())+df["Book-Rating"].apply(lambda x:x.split())
df=df.drop(columns=["Book-Author","Year-Of-Publication", "Publisher", "Book-Rating"])
# df["tag"]=df["tag"].apply(lambda x:x.split())
df
```

```python
df["tag"]=df["tag"].apply(lambda x:" ".join(x))
df["tag"]=df["tag"].str.lower()
df
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english') # we need 5000 freq. for prediction and we r removing the stop words of english lang.
```

```python
vector=cv.fit_transform(df['tag']).toarray()
vector
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
sorted(enumerate(similarity[0]),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(book_name):
    book_index=df[df['Book_Title']==book_name].index[0]
    l=sorted(enumerate(similarity[book_index]),reverse=True,key=lambda x:x[1])[1:7]
    for i in l:
        print(df.iloc[i[0]].Book_Title)
```
